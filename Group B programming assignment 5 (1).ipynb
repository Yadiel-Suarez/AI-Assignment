II. Concepts

To develop an effective adversarial game, it is essential to first understand the concept of adversarial search. Adversarial search is defined as “a well-suited approach in a competitive environment, where two or more agents have conflicting goals” [1]. The classic game of dominoes aligns well with this definition, as it inherently involves two or more players competing with opposing objectives. In this context, one of the players is represented by an AI agent. The role of this agent is to “make optimal decisions while anticipating the actions of an opponent with their opposing objectives” [1].

Adversarial search encompasses two key components: game-playing and decision-making. Game-playing refers to the ability to represent “the state of a game in a straightforward approach and the agents are limited to a small number of actions whose effects are governed by precise rules” [1]. This definition directly relates to the nature of dominoes, where players operate under a fixed set of rules and a limited number of valid moves. Consequently, the game-playing component in adversarial search is crucial in replicating the structure and legality of gameplay in dominoes.

In parallel, the decision-making aspect of adversarial search enables an AI to "find the best possible move or strategy for a player in a competitive environment against one or more components...[having] adaptive decision-making throughout the game" [1]. Dominoes exemplifies this principle well, as players must constantly evaluate the board state and make real-time decisions based on partial information. Furthermore, the game requires the application of a heuristic evaluation function, which allows the AI to assign numerical values to each game state. This function evaluates states “based on piece positions, material advantages, and board control” [1], allowing the agent to assess potential strategies effectively.

To implement adversarial logic into the AI’s decision-making process, various strategies were considered. The first approach explored was the technique of averaging over clairvoyance, in which “the agent can compute the expectiminimax, and choose the move that yields the highest value” [2]. This approach “assumes that every future state will automatically be one of perfect knowledge” [2]. However, due to the inherent uncertainty and incomplete information present in a typical dominoes match, this technique was found to be incompatible. The game requires players to make real-time decisions based on limited information, which contradicts the clairvoyant assumption of full visibility of future states.

Another technique that was evaluated but ultimately dismissed was the Monte Carlo Simulation. This method is defined as “a large representative sampling of stochastic system trajectories..., where one gains insight to the system through collection and analysis of trajectory data” [3]. While this simulation method can provide value in terms of analyzing potential outcomes, it was deemed more suitable as a supplementary tool rather than a foundational strategy. This is because the sampling necessary to simulate meaningful trajectories is not easily applicable to a game like dominoes, where decisions are primarily based on immediate circumstances and strategic foresight rather than long-term probability distributions.

Although dominoes is predominantly a deterministic game—where outcomes are governed by clear and structured rules—it also contains a stochastic element at the initial state, due to the random distribution of tiles among players. This initial randomness aligns with the concept of stochastic games, defined as “multi-agent games where players take turns making decisions, influencing the game's current state and future states...[It involves] randomness in determining transitions between states, making decision-making more complex” [4]. The core challenge here is for players to “interact and must make decisions under uncertainty, which directly influences both immediate and long-term results” [4].

Given the blend of deterministic and stochastic features in dominoes, the most appropriate adversarial search strategy identified for this game was the partially observable game technique. Under this model, the AI agent constructs decisions based on limited information—specifically, what is visible or known at a particular point in the game. Partial observability is defined as having “access to partial information about the environment...[even if they exist] aspects of the environment may be hidden or unknown, forcing the agent to make decisions with uncertainty” [5]. In dominoes, players are only aware of their own tiles and the tiles played on the board; the opponent's hand remains unknown. This layer of hidden information significantly increases the complexity of the problem, as “the agent must rely on predictions or assumptions about hidden aspects, making the problem more complex” [5].

Overall, the game of dominoes presents a compelling environment for adversarial search due to its strategic depth, structured rules, and inherent uncertainty. After evaluating several adversarial techniques, the partially observable approach emerged as the most appropriate model. This method allows the AI agent to simulate human-like reasoning and decision-making, effectively adapting its strategy despite incomplete information—an essential feature for replicating intelligent adversarial gameplay.

IV. Analysis

The implementation of the domino game required careful consideration of two key aspects: understanding the intrinsic rules of the game and identifying an appropriate artificial intelligence (AI) technique to support gameplay. Before delving into code-specific details, it is essential to outline the core functionality and rules. This implementation is modeled after the traditional Puerto Rican version of dominoes, adapted for a simplified two-player format: one human participant and one AI-controlled agent. The game can conclude in one of three possible outcomes: a win for the participant, a loss to the AI agent, or a tie. It is a turn-based game where a central domino initiates play, and players alternate placing dominoes that match the numbers at either end of the growing chain. Dominoes may be placed on either side, provided a matching number is present. If a player cannot play a valid domino from their hand, they must draw from a stockpile. A player wins when all dominoes in their hand are successfully played. Conversely, the player loses if the AI plays all of its dominoes first. A tie occurs when neither the participant nor the AI can make a valid move, the stockpile is depleted, and both players still retain dominoes in hand.

Several notable characteristics define the nature of this game. It is adversarial, as the two players have directly opposing goals. It is partially observable, given that neither player has access to the opponent's hand. The game also exhibits both stochastic and deterministic elements: the initial shuffle introduces randomness, but once hands are dealt, player decisions and outcomes become deterministic. Furthermore, uncertainty persists throughout the game due to the hidden states of the opponent’s hand and the remaining stockpile. Lastly, dominoes is a sequential game, where each decision directly influences future states and potential outcomes.

The selection of an appropriate AI technique was a central component of the design process. Several methods were evaluated and ultimately excluded. One such method, averaging over clairvoyance, was dismissed early due to its underlying assumption of full knowledge of hidden states and the requirement to average over multiple fully observable playthroughs. While theoretically sound, it proved impractical within the constraints and goals of this domino implementation.

Another technique considered but not adopted was Monte Carlo simulation. This method is powerful for simulating numerous potential game states and could significantly enhance agent performance by evaluating sampled outcomes and estimating optimal plays. However, the decision to prioritize the simplicity and cultural essence of Puerto Rican dominoes over increased computational complexity led to its exclusion. In Puerto Rican culture, the appeal of dominoes lies not in its difficulty, but in its strategic and logical foundations. It is a game celebrated across generations, valued for its accessibility, and often played during family gatherings—a central aspect of Puerto Rican social life.

The stochastic game approach was also reviewed. Although applicable to the game’s random initial state, it was deemed unsuitable beyond that point, as the gameplay becomes largely deterministic. Given the dominance of deterministic decision-making after the shuffle, the stochastic game model was not a practical fit.

The most appropriate framework for modeling this game proved to be the partially observable game technique. Puerto Rican dominoes inherently operate under partial observability due to the hidden nature of the players’ hands and the stockpile, even without the involvement of an AI. With the addition of an AI agent, this property remains unchanged—the agent lacks access to the opponent’s hand and must make decisions based solely on the observable state (i.e., the domino chain) and inferred knowledge. The AI employs heuristic reasoning, particularly by evaluating the frequency of visible domino values to determine its moves. This mimics probabilistic reasoning in a partially observable environment. The agent, therefore, operates under uncertainty and limited information, much like a human opponent.

As previously mentioned, portions of the code were adapted from open-source contributions, particularly from various GitHub repositories. All borrowed logic and code snippets are credited accordingly in the references. [7] The functionality closely emulates the real-world rules and dynamics of Puerto Rican dominoes. Furthermore, the codebase is thoroughly documented and commented to enhance readability and comprehension.


```python
from random import shuffle
from itertools import combinations_with_replacement, chain
from collections import Counter

# --- Game Logic Functions ---

def turno_per_mano(mano_input, fichas):
    turno = int(mano_input)
    if turno == 0:
        if stock_fichas:
            print("Te tocó robar!")
            fichas.append(stock_fichas.pop())
        return

    ficha = fichas[abs(turno) - 1]

    if turno > 0:
        if ficha[1] == double_chain[-1][-1]:
            ficha.reverse()
        double_chain.append(ficha)
    else:
        if ficha[0] == double_chain[0][0]:
            ficha.reverse()
        double_chain.insert(0, ficha)

    if ficha == [6, 6]:
        print("Sacó la caja de dientes! (double six)")

    fichas.pop(abs(turno) - 1)

# --- Game Setup ---

dominoes = [list(d) for d in combinations_with_replacement(range(7), 2)]
assert len(dominoes) == 28, "Domino set must have exactly 28 tiles!"
shuffle(dominoes)

player_fichas = dominoes[:7]
computer_fichas = dominoes[7:14]
stock_fichas = dominoes[14:]

doubles = [d for d in computer_fichas + player_fichas if d[0] == d[1]]
double_chain = [max(doubles)] if doubles else []
if double_chain[0] in computer_fichas:
    computer_fichas.remove(double_chain[0])
    num_of_mano = 1  # player's turn
else:
    player_fichas.remove(double_chain[0])
    num_of_mano = 0  # computer's turn

# --- Slangs & Intro ---
# input("\nLet's learn the lingo...When you are done press 'ENTER' to start game.")
# print("\nLet's learn the lingo...When you are done press 'ENTER' to start game.")

print("\nFicha: single domino")
print("\nMano: single round in a match. A dominoes match is made of several games (manos).")
print("\nTe toca jugar: It's your turn")
print("\nChuchazo: When the winning bone is the double-blank (“la Chucha!”)")
print("\nCapicú: When the winning bone can play on either open end of the layout. Doesn’t apply when the winning bone is a double.")
print("\nTrancado: A blocked game. The last bone played before the game got blocked is called the “Tranque“")
print("\nCaja de Diente: Double six dominoes")
print("\nTe tocó robar: Took dominoes from stock pile")

# input()
input("\nLet's learn the lingo...When you are done press 'ENTER' to start game.")



# --- Game Loop ---

while True:
    print('=' * 70)
    print('Stock size:', len(stock_fichas))
    print('Computer fichas:', len(computer_fichas), '\n')
    if len(double_chain) <= 6:
        print(*double_chain, '\n', sep='')
    else:
        print(*double_chain[:3], '...', *double_chain[-3:], '\n', sep='')

    print("Your fichas:")
    for i, p in enumerate(player_fichas, start=1):
        print(f"{i}: {p}")

    # Check for victory or draw
    if not player_fichas:
        mano_final = double_chain[-1]
        ends = [double_chain[0][0], double_chain[-1][-1]]
        if mano_final == [0, 0]:
            print("\nStatus: The game is over. You won! Eso fue un chuchazo!")
        elif ends[0] == ends[1]:
            print("\nCapicú! Status: The game is over. You won!")
        else:
            print("\nStatus: The game is over. You won!")
        break

    if not computer_fichas:
        print("\nStatus: The game is over. The computer won!")
        break

    ends = [double_chain[0][0], double_chain[-1][-1]]
    fichas_remaining = list(chain(*player_fichas, *computer_fichas))
    if not stock_fichas and not any(num in ends for num in fichas_remaining):
        print("\nStatus: The game is over. Trancado!")
        break

    # Player move
    if num_of_mano % 2 == 0:
        print("\nTe toca jugar! Status: It's your turn to make a move. Enter your command.")
        mano_input = input()
        if mano_input.lstrip("-").isdigit():
            manos = int(mano_input)
            if -len(player_fichas) <= manos <= len(player_fichas):
                if manos == 0:
                    turno_per_mano(mano_input, player_fichas)
                    num_of_mano += 1
                else:
                    piece = player_fichas[abs(manos) - 1]
                    if (manos > 0 and ends[1] in piece) or (manos < 0 and ends[0] in piece):
                        turno_per_mano(mano_input, player_fichas)
                        num_of_mano += 1
                    else:
                        print("Illegal move. Please try again.")
            else:
                print("Invalid input. Please try again.")
        else:
            print("Invalid input. Please try again.")
    # Computer move
    else:
        print("\nThe computer le toca jugar. Status: Computer's mano. Press Enter to continue...")
        input()
        count = Counter(chain(*computer_fichas, *double_chain))
        scored_fichas = sorted(computer_fichas, key=lambda x: count[x[0]] + count[x[1]], reverse=True)

        for piece in scored_fichas:
            i = computer_fichas.index(piece)
            if ends[1] in piece:
                turno_per_mano(str(i + 1), computer_fichas)
                break
            elif ends[0] in piece:
                turno_per_mano(str(-i - 1), computer_fichas)
                break
        else:
            turno_per_mano('0', computer_fichas)
        num_of_mano += 1
```
